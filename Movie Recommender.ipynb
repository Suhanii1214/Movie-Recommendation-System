```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv(r"C:\Users\hp\Downloads\movies_dataset\movies.csv")
credits = pd.read_csv(r"C:\Users\hp\Downloads\movies_dataset\credits.csv")
```

```python
movies
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies = movies[['movie_id', 'overview', 'title', 'genres', 'cast', 'crew', 'keywords']]
```

```python
movies
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.iloc[0].genres
```

```python
#Extracting names of all genres ffor each movie
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['genres']
```

```python
movies
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert_cast(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter<3:
            L.append(i['name'])
            counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.iloc[0].crew
```

```python
def convert_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
    
    return L
```

```python
movies['crew']=movies['crew'].apply(convert_crew)
```

```python
movies
```

```python
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies
```

```python
movies['tags'] = movies['overview'] + movies['keywords'] + movies['cast'] + movies['crew'] + movies['genres']
```

```python
movies
```

```python
new_movies = movies[['movie_id', 'title', 'tags']]
```

```python
new_movies
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x))
```

```python
new_movies
```

```python
new_movies.iloc[0].tags
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x: x.lower())
```

```python
new_movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stems(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_movies['tags'] = new_movies['tags'].apply(stems)
```

```python
new_movies
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_movies[new_movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
        
recommend('Batman')
```
